**Create Graph**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
```

**Adding nodes to the graph**

1. Add a single node

```python
G.add_node('A')

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

2. Add multiple nodes from a list

```python
G.add_nodes_from(['B', 'C', 'D', 'E'])

nx.draw(G, with_labels=True)
```

**Adding edges to the graph**

1. Add a single edge (nodes 'B' and 'C' already exist)

```python
G.add_edge('B', 'C')
nx.draw(G, with_labels=True)
```

3. Add multiple edges from a list of tuples

```python
G.add_edges_from([('A', 'B'), ('C', 'D'), ('A', 'D'), ('A', 'A'), ('C', 'A')])
nx.draw(G, with_labels=True)
```

**Fetch Graph Information**

1. Nodes and Edges in a Graph

```python
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

2. No of Nodes and Edges in a Graph

```python
print("No of Nodes:", G.number_of_nodes())
print("No of Edges:", G.number_of_edges())
```

3. List of Nodes and Edges in a graph

```python
print("List of nodes:", list(G.nodes()))
print("List of edges:", list(G.edges()))
```

4. Degree of any vertex of graph

```python
print("Degree of A:", G.degree['A'])
```

**Removing nodes from a graph**

```python
G = nx.wheel_graph(5)
nx.draw(G, with_labels=True)
```

```python
print("Original Graph:")
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

1. Remove node n.

  Removes the node n and all adjacent edges. Attempting to remove a nonexistent node will raise an exception.

```python
G.remove_node(0)
nx.draw(G, with_labels=True)
```

2. Graph.remove_nodes_from(nodes)

    Remove multiple nodes.

```python
G.remove_nodes_from([1, 2])
nx.draw(G, with_labels=True)
```
