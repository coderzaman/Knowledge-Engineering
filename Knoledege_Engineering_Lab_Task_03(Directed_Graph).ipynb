```python
import networkx as nx
import matplotlib.pyplot as plt
```

**Task 2.1: Create Different Types of Graphs**

Create two directed graphs: one with 7 nodes and 10 edges, and another with 6 nodes and 14 edges.


  * 7 nodes and 10 edges

```python
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6,7])
G.add_edges_from([(1,2),(1,3),(2,4),(4,5),(4,6),(5,7),(6,7),(5,2),(6,1),(2,1)])

p = nx.spring_layout(G, seed=31)
nx.draw(G, with_labels=True, pos=p)
plt.show()
```

  * 6 nodes and 14 edges

```python
H = nx.DiGraph()
H.add_nodes_from([1,2,3,4,5,6])
H.add_edges_from([(1,2),(1,3),(2,4),(4,5),(4,6),(5,3),(6,2),(3,2),(6,6),(1,4),(5,2),(6,1),(2,1), (6,3)])

p = nx.spring_layout(H, seed=31)
nx.draw(H, with_labels=True, pos=p)
plt.show()
```

**Task 2.3: Self-Loop and Multi-Edge Handling**

```python
Loop_G = nx.DiGraph()
Loop_G.add_nodes_from(['A','B','C','D'])
Loop_G.add_edges_from([('A','B'),('B','C'),('C','D'),('D','D')])
p = nx.spring_layout(Loop_G, seed=31)
nx.draw(Loop_G, with_labels=True,pos=p)
plt.show()
```

**Task 2.5: Component Manipulation**

* Start with a connected graph
* Systematically remove nodes/edges to create disconnected components
* Track how connectivity changes after each removal

```python
Man_G = nx.DiGraph()
Man_G.add_nodes_from(['1','2','3','4','5'])
Man_G.add_edges_from([('1','2'),('2','3'),('3','4'),('4','5'),('5','1')])
p = nx.spring_layout(Man_G, seed=31)
nx.draw(Man_G, with_labels=True, pos=p)
plt.show()
```

1. Remove Node

```python
Man_G.remove_node('1')
nx.draw(Man_G, with_labels=True, pos =p)
plt.show()
```

2. Remove Edge

```python
Man_G.remove_edge('2','3')
p = nx.spring_layout(Man_G, seed=31)
nx.draw(Man_G, with_labels=True,pos=p)
plt.show()
```

Task 2.6: Connectivity Restoration

Connectivity Restoration Start with a disconnected graph, find the minimum number of edges needed to make it connected

```python
Con_Grap = Man_G.copy()
Con_Grap.add_node('1')
p = nx.spring_layout(Con_Grap, seed=31)
nx.draw(Con_Grap, with_labels=True, pos=p)
plt.show()
```

**Approach For Solve this Problem**
* Find (k): First, count the number of disconnected components.

* Calculate (k−1): The minimum number of edges needed is always one less than the number of components with nx.number_weakly_connected_components(Graph) Function.

1. Find (k): First, count the number of disconnected components.

```python
k = nx.number_weakly_connected_components(Con_Grap)
print(f"Number of connected components: {k}")
```

2. The minimum number of edges needed

```python
print(f"Minimum number of edges needed: {k-1}")
nx.draw(Con_Grap, with_labels=True,pos=p)
plt.show()
```

**Connect this graph with adding edges**

```python
Con_Grap.add_edges_from([('2','3'),('2','1')])
p = nx.spring_layout(Con_Grap, seed=31)
nx.draw(Con_Grap, with_labels=True, pos=p)
plt.show()
```
