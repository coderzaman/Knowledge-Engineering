**Task 1.5: Component Manipulation**

* Start with a connected graph
* Systematically remove nodes/edges to create disconnected components
* Track how connectivity changes after each removal

```python
Man_G = nx.Graph()
Man_G.add_nodes_from(['1','2','3','4','5'])
Man_G.add_edges_from([('1','2'),('2','3'),('3','4'),('4','5'),('5','1')])
nx.draw(Man_G, with_labels=True)
plt.show()
```

1. Remove Node

```python
Man_G.remove_node('1')
nx.draw(Man_G, with_labels=True)
plt.show()
```

2. Remove Edge

```python
Man_G.remove_edge('3','2')
nx.draw(Man_G, with_labels=True)
plt.show()
```

**Task 1.6: Connectivity Restoration**

Connectivity Restoration Start with a disconnected graph, find the minimum number of edges needed to make it connected

```python
Con_Grap = Man_G.copy()
nx.draw(Con_Grap, with_labels=True)
plt.show()
```

**Add Edges**

```python
Con_Grap.add_edge('3','2')
nx.draw(Con_Grap, with_labels=True)
plt.show()
```
